```python
import yfinance as yf

df_gold = yf.download('GC=F', interval='1d', period='10y')
df_gold.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in df_gold.columns]
df_gold.columns = [c.split('_')[0].lower() for c in df_gold.columns]
df_gold = df_gold[['open', 'high', 'low', 'close', 'volume']]

df_eur = yf.download('EURUSD=X', interval='1d', period='10y')
df_eur.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in df_eur.columns]
df_eur.columns = [c.split('_')[0].lower() for c in df_eur.columns]
df_eur = df_eur[['open', 'high', 'low', 'close', 'volume']]

df_cl = yf.download('CL=F', interval='1d', period='10y')
df_cl.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in df_cl.columns]
df_cl.columns = [c.split('_')[0].lower() for c in df_cl.columns]
df_cl = df_cl = df_cl[['open', 'high', 'low', 'close', 'volume']]

print(f"Gold bars: {len(df_gold)}")
print(f"EURUSD bars: {len(df_eur)}")
print(f"CL bars: {len(df_cl)}")
```

```python
import pandas as pd
from datetime import time

system_test = VPORBSystem(instrument= 'GC', equity=2000)
window = df_gold.iloc[:150]

vp = system_test.vp_engine.build(window)
orb = system_test.orb_engine.build_orb(window)
regime = system_test.regime_filter.classify(window)

print(f"Regime: {regime}")
print(f"ORB valid: {orb.valid}")
print(f"POC: {vp.poc}")
print(f"VAH: {vp.vah}")
print(f"VAL: {vp.val}")
print(f"Current price: {window.iloc[-1]['close']}")

signal = system_test.signal_engine.evaluate(
    window, vp, orb, regime, 2000
)
print(f"Signal: {signal}")
```
